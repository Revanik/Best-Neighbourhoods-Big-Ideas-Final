```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
speedHumps = pd.read_csv("CSV-Files/SpeedHumps.csv")
speedHumps.head()
```

Printing a section of the dataset explained how the data table was organized

The created dataframe needed to be reorganized, with renamed columns for interpretability and for the sake of the graph

```python
newHumps = pd.DataFrame(speedHumps['neighborhood'].value_counts().reset_index())
newHumps = newHumps.rename(columns = {'neighborhood':'Neighborhoods','count':'Count'})
newHumps
```

```python
newHumps.plot(kind='bar',x='Neighborhoods',y='Count',title = 'Top 18')

plt.tight_layout()
```

According to Allegheny County's dataset regarding the number of speed humps in a specified neighborhood, Squirrel Hill South is the neighborhood that has the most, at 18 speed humps, followed by Carrick at 9 and North Oakland at 6. While this may seem like an incovenience for drivers, this metric represents a symbol of safety for pedestrians and drivers alike, as it requires cars to slow down in the area where there is the most foot traffic. It is very likely that population size and type in the given neighborhoods may correspond with the number of speed humps. Overall, since Squirrel Hill South has the most speed humps, it is the best neighborhood for pedestrian safety.
