Filter out all inactive crosswalks.

```python
walksActive = walks[walks["inactive"] != "t"]
```

```python
crosswalkCounts = pd.DataFrame(walksActive['neighborhood'].value_counts())
crosswalkCounts.head()
```

```python
crosswalkCounts = crosswalkCounts.sort_values(by='count', ascending=False)
crosswalkCounts
```
