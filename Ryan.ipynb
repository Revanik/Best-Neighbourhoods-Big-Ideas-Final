```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
walks = pd.read_csv("CSV-Files/Crosswalks.csv")
walks.head()
```

Example of what the data looks like in the csv file

Filter out all inactive crosswalks and count 

```python
walksActive = walks[walks["inactive"] != "t"]
crosswalkCounts = pd.DataFrame(walksActive['neighborhood'].value_counts())
crosswalkCounts.head()
crosswalkCounts = crosswalkCounts.sort_values(by='count', ascending=False)
crosswalkCounts
```

```python
crosswalkCounts = crosswalkCounts.iloc[1:]
crosswalkCounts
```

```python
crosswalkCounts.plot()
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
